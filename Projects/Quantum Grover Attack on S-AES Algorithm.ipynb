## 1. Introduction
__In cryptography,__ the asymmetric cryptographic algorithms are mainly used to exchange the secret encryption key which is used by a symmetric cryptographic algorithm.



#### Symmetric Encryption Examples:
The symmetric encryption technique is used worldwide for data security and transportation.

* Advanced Encryption Standard (AES)\
AES encryption is one of the most common encryption standards mostly used in internet networks. It uses block ciphers of 128, 192, or 256 bits to encrypt and decrypt the data. The encryption procedure of AES is so complex that it is nearly impossible to crack this technique. 

#### Benefits of Symmetric Cryptography:
* Swift Processing\
The symmetric encryption technique uses shorter key lengths as compared to asymmetric encryption, so they are much faster to execute.

* Higher Security\
Symmetric encryption algorithms such as AES are highly secure and nearly impossible to crack.

* Ability to Process Large Files\
One of the significant benefits of symmetric encryption is that it can work with large-size files more easily and faster.

#### Disadvantages of Symmetric Encryption:

* No Secure Method for Key Distribution\
In this method, an insecure channel compromises the security of the process by transmitting the keys.

* No Digital Signatures\
Digital signatures ensure that the message comes from a secure source. Unfortunately, symmetric encryption does not have any method for enabling digital signatures.

#### Asymmetric Encryption Examples

* Rivest Shamir Adleman (RSA)\
RSA is one of the most prominent examples of this encryption technique. It generates a public key for the encryption part by multiplying two large-sized, random prime numbers together. It also generates the private key needed for decrypting the data from the same prime numbers.

## 2. Case study:
A simplified version of the Advanced Encryption Standard (S-AES); using Grover's Search Algorithm as a subroutine for performing an attack on it to find the symmetric KEY used in encryption.

__In this paper,__ Almazrooie et al. show that a quantum ciruit of a polynomial number of basic quantum gates can be designed for a Block Cipher by using the S-AES.

__AES Algorithm__ is a multi-round substitution/permutation cipher. It is considered quantum-resistant for some key sizes but vulnerable to a Grover’s search attack for smaller key sizes.

__AES128 is vulnerable to a Grover search attack.__

## Problem Definition:
* Given a set of plain message and ciphertext pairs we can define our optimization function,\
  i.e. find the key which is consistent with our plain/ciphertext pairs.\
  The unitary implementation of this function will serve as the Oracle.

* Once a quantum adversary implemented a block cipher as a complete reversible quantum circuit, that quantum circuit can be integrated as a function within an Oracle or Black-box.
* This Black-box is a sort of Boolean function which is a basic building
component in other quantum algorithms such as Grover’salgorithm.

## The Simplified-AES:
* is a small variant of the AES.
* has the exact similar structure as AES with small parameters.
* The plaintext block size in S-AES is 16 bits and the key size is also 16 bits instead of 128.
* is a Nibble-oriented algorithm compared to a Byte-oriented algorithm in AES (one Nibble is four bits).
* have been used for the purposes of classical cryptanalysis.
* The state array of S-AES is:
  $$ \overbrace{b_0 b_1 b_2 b_3}^{S_0} \overbrace{b_4 b_5 b_6 b_7}^{S_1} \overbrace{b_8 b_9 b_{10} b_{11}}^{S_2} \overbrace{b_{12} b_{13} b_{14} b_{15}}^{S_3} \Longrightarrow  State = \begin{pmatrix} S_0 & S_2 \\ S_1 & S_3 \end{pmatrix} $$
  where $ b_i $ is an input bit and $ S_i $ is one Nibble.

* this S-AES consists of two rounds as illustrated in the Figure:
  


#### The main four steps of S-AES are:
* Nibbles substitution (__SubNibbles__) corresponds __Substitute Byets__ in AES ,
* rows shifting (__ShiftRows__),
* columns mixing (__MixColumns__),
* and adding round key (__AddKey__).

Similarly to AES, MixColumns is excluded from the last round in S-AES, and the subkeys generation algorithm resamples also the one in AES.



## 3. Grover's Algorithm
Given an unstructured database of N
elements, find the element a ∈ N. This can be modeled as a
function $ f: \{0, 1\}^n → \{0, 1\}^n$ , where the space $N = 2^n$ , for any $x ∈ \{0, 1\}^n$
$$ f(x) = \begin{cases} 1 & if \ x= a \ (a\ solution) \\ 0 & otherwise\ (not\ a\ solution) \end{cases} $$

One can find (with probability $ > 1/2 $) a specific item __- single match -__ within a randomly ordered database of $N$ items using $ O(\sqrt{N}) $ operations. By contrast, a classical computer would require $ O(N) $ operations to achieve this. Therefore, Grover’s algorithm provides a quadratic speedup over an optimal classical algorithm. It has also been shown that Grover’s algorithm is optimal in the sense that no quantum Turing machine can do this in less than $ O(\sqrt{N}) $ operations.
 ***
__Grover's Algorithm:__\
__Input:__ An unstructured set $ N = {a_1, a_2, ..., a_n} $\
__Output:__ $ a_i \in N $\
__1 Step 1: Initialization of the quantum register:__\
__2__ all the quabits $ x^{\otimes^n} $ to $\ket0 $ state and the oracle qubit $q$ to $\ket1 $ state:\
__3__ $\ket{\psi_0} = \ket0^{\otimes^n} \ket1 $\
__4 Step 2: Put the register in an uniformly distributed superposition:__\
__5__ apply __H__ Hadamarad gate:\
__6__ $ \ket{\psi_1} = \frac{1}{\sqrt{N}} \sum^{N-1}_{i=0} \ket{i} \otimes \frac{\ket0 - \ket1}{\sqrt2}$ \
__7 Step 3: Apply Grover iterations:__\
__8 for__ $ 2^{n/2} $ times, __do__\
    __a - Apply the oracle:__\
    $\ket{x} \overset{O}{\longrightarrow} (-1)^{f(x)} \ket{x} $\
    __b - Perform Grover Operator (inversion about the mean):__\
    i. Apply $ H^{\otimes{n}} $\
    ii. Conditionally shift phase\
    iii. Apply $ H^{\otimes{n}} $\
__9 step 4: Measure the quantum register__

## 4. The Simplified-AES
The entire design is mainly structured with a consideration of utilizing the lowest possible number of qubits even though that could lead to a growing number quantum gates needed.


#### 4.1 Nibbles Substitution:
provides the non-linear diffusion in S-AES. It takes a Nibble (a) in the form of a 4-bit column vector as an input and substitutes it with a different Nibble (b) according the following steps:
* Calculating the multiplicative inverse using __square-multiply__ approach.
  The finite field of S-AES is $GF(2^4)$ with irreducible polynomial $ x^4 + x + 1 $.
* Calculating the affine transformation.
  
  
   

```python
# Define the multiplier quantum circuit of the finite field
# The inputs a, b vectors, and the output is c vector, such that c = a x b.

def multiplier():    # c: workspace qubits initially 0
    a = QuantumRegister(4, 'a')
    b = QuantumRegister(4, 'b')
    c = QuantumRegister(4, 'c')
    qc = QuantumCircuit(c, b, a, name="M")
    
    qc.ccx(a[3], b[1], c[0])
    qc.ccx(a[2], b[2], c[0])
    qc.ccx(a[1], b[3], c[0])
    qc.ccx(a[3], b[2], c[1])
    qc.ccx(a[2], b[3], c[1])
    qc.ccx(a[3], b[3], c[2])
    qc.cx(c[2], c[3])
    qc.cx(c[1], c[2])
    qc.cx(c[0], c[1])
    qc.ccx(a[3], b[0], c[3])
    qc.ccx(a[2], b[1], c[3])
    qc.ccx(a[1], b[2], c[3])
    qc.ccx(a[0], b[3], c[3])
    qc.ccx(a[2], b[0], c[2])
    qc.ccx(a[1], b[1], c[2])
    qc.ccx(a[0], b[2], c[2])
    qc.ccx(a[1], b[0], c[1])
    qc.ccx(a[0], b[1], c[1])
    qc.ccx(a[0], b[0], c[0])
    return qc

multiplier().draw('mpl', reverse_bits=True)
```

```python
# Define the squarer circuit of the finite field such that ~a = a x a
def xsquare():
    N = QuantumRegister(4, 'a')
    qc = QuantumCircuit(N, name="x^2")
    qc.cx(N[2], N[0])
    qc.cx(N[2], N[1])
    qc.cx(N[1], N[2])
    qc.cx(N[2], N[1])
    qc.cx(N[3], N[2])
    return qc
    
display(xsquare().draw('mpl', reverse_bits=True))
display(Operator(xsquare()).draw('latex', max_size=16))
```

```python
# Define the quantum circuit of the affine transformation of S-AES
def FTrans():
    N = QuantumRegister(4, 'a')
    qc = QuantumCircuit(N, name="F")
    qc.cx(N[1], N[0])
    qc.cx(N[2], N[0])
    qc.cx(N[2], N[1])
    qc.cx(N[3], N[1])
    qc.cx(N[2], N[3])
    qc.cx(N[0], N[3])
    qc.cx(N[1], N[2])
    qc.cx(N[0], N[2])
    qc.x([N[0], N[3]])
    return qc

display(FTrans().draw('mpl', reverse_bits=True))
display(Operator(FTrans()).draw('latex', max_size=16))
```

```python
# Define a Quantum Circuit to Substitute Nibbles

def subNibbles():
    # define the Quantum registers and the Quantum Circuit
    N0 = QuantumRegister(4, 'N0')
    N1 = QuantumRegister(4, 'N1')
    N2 = QuantumRegister(4, 'N2')
    N3 = QuantumRegister(4, 'N3')
    qc = QuantumCircuit(N0, N1, N2, N3, name="SN")

    # Apply the sub gates
    qc.append(xsquare(), list(N0))
    qc.append(MCMT("x", 4, 4), [*N0, *N1])        
    qc.append(xsquare(), list(N0))
    qc.append(multiplier(), [*N0, *N1, *N2])
    qc.append(xsquare(), list(N0))
    qc.append(multiplier(), [*N0, *N2, *N3])
    
    qc.append(xsquare().inverse(), list(N0))
    qc.append(multiplier().inverse(), [*N0, *N1, *N2])
    
    qc.append(xsquare().inverse(), list(N0))
    qc.append(MCMT("x", 4, 4), [*N0, *N1])        
    qc.append(xsquare().inverse(), list(N0))
    qc.append(FTrans(), list(N3))

    return qc
    
qc = QuantumCircuit(*N[3::-1])
qc.append(subNibbles(), [*N[0],*N[1],*N[2],*N[3]])
qc.decompose().draw('mpl', reverse_bits=True)
```

#### 4.3 Columns Mixing:
The MixColumns transformation in S-AES is expressed as the following matrix multiplication:





```python
# Define a Quantum Circuit to Mix Columns

def mixColumns():
    qc = QuantumCircuit(8, name="MC")
    qc.cx(7,1)
    qc.cx(6,0)
    qc.cx(6,1)
    qc.cx(4,2)
    qc.cx(5,3)
    qc.cx(7,2)
    qc.cx(7,5)
    qc.cx(7,6)
    qc.cx(6,7)
    qc.cx(7,6)
    qc.cx(5,6)
    qc.cx(6,5)
    qc.cx(4,5)
    qc.cx(5,4)
    qc.cx(3,5)
    qc.cx(2,5)
    qc.cx(3,6)
    qc.cx(2,4)
    qc.cx(1,7)
    qc.cx(0,6)
    return qc

display(mixColumns().draw('mpl', reverse_bits=True))
# display(Operator(mixColumns()).draw('latex', max_size=16))
```

```python
# Define the Plain Text quantum circuit:
plaintext = "1101011100101000"

def plain(ptxt):
    qc = QuantumCircuit(16, name="PTXT")
    for i in range(len(ptxt)):
        if ptxt[len(ptxt)-i-1] == '1':
            qc.x(i)
    return qc

plain(plaintext).draw('mpl', reverse_bits=True)
```

```python
# Define the cipher quantum circuit:
ciphertext = "0010010011101100"
def cipher(ctxt):
    qc = QuantumCircuit(16, name="C*")
    for i in range(len(ctxt)):
        if ctxt[len(ctxt)-i-1] == '1':
            qc.x(i)
    return qc

cipher(ciphertext).draw('mpl', reverse_bits=True)
```

```python
plaintext = "1101011100101000"
ciphertext = "0010010011101100"
key = "0100101011110101" 

# Define the proposed quantum circuit of the Simplified-AES
def SAES():
    # Define a state of 64 qubit devided into 16 Nibbles, each one includes 4 qubits
    N = []
    for n in range(16):
        n = QuantumRegister(4, f'N{n}')    # each Nibbles includes 4 qubits
        N.append(n)

    # Define 8 ancillay qubit to be used in the S-box in subNibbles circuit
    a1 = QuantumRegister(4, 'a1')
    a2 = QuantumRegister(4, 'a2')
    
    # Define the Quantum Circuit)
    qc = QuantumCircuit(*N[::-1], a1, a2, name="P*SAES")
    qc.barrier()
    qc.append(plain(plaintext), [*N[3],*N[2], *N[1], *N[0]])
    qc.append(subNibbles(), [*N[0], *a1, *a2, *N[10]])
    qc.append(subNibbles(), [*N[1], *a1, *a2, *N[9]])
    qc.append(subNibbles(), [*N[2], *a1, *a2, *N[8]])
    qc.append(subNibbles(), [*N[3], *a1, *a2, *N[11]])

    qc.barrier()
    qc.append(plain(plaintext), [*N[3],*N[2], *N[1], *N[0]])
    qc.append(mixColumns(), [*N[9],*N[8]])
    qc.append(mixColumns(), [*N[11],*N[10]])

    qc.barrier()
    qc.append(subNibbles(), [*N[2], *a1, *a2, *N[5]])
    qc.append(subNibbles(), [*N[3], *a1, *a2, *N[4]])
    # qc.append(rc(), [*N[5]])

    qc.barrier()
    qc.append(MCMT("x", 4, 4), [*N[0], *N[4]])       
    qc.append(MCMT("x", 4, 4), [*N[1], *N[5]])       
    qc.append(MCMT("x", 4, 4), [*N[4], *N[8]])       
        
    
    qc.barrier()
    qc.append(MCMT("x", 4, 4), [*N[2], *N[4]])       
    qc.append(MCMT("x", 4, 4), [*N[3], *N[5]])       
    qc.append(MCMT("x", 4, 4), [*N[4], *N[10]])
    qc.append(MCMT("x", 4, 4), [*N[5], *N[11]])       
    
    qc.barrier()
    qc.append(subNibbles(), [*N[8], *a1, *a2, *N[14]])
    qc.append(subNibbles(), [*N[9], *a1, *a2, *N[13]])
    qc.append(subNibbles(), [*N[10], *a1, *a2, *N[12]])
    qc.append(subNibbles(), [*N[11], *a1, *a2, *N[15]])
    qc.append(subNibbles(), [*N[4], *a1, *a2, *N[7]])
    qc.append(subNibbles(), [*N[5], *a1, *a2, *N[6]])
    # qc.append(rc(), [*N[5]])
    qc.append(MCMT("x", 4, 4), [*N[3], *N[5]]) 
    qc.append(MCMT("x", 4, 4), [*N[2], *N[4]])       

    qc.barrier()
    qc.append(MCMT("x", 4, 4), [*N[4], *N[6]])      
    qc.append(MCMT("x", 4, 4), [*N[5], *N[7]])      
    qc.append(MCMT("x", 4, 4), [*N[6], *N[12]])      
    qc.append(MCMT("x", 4, 4), [*N[7], *N[13]])      
    qc.append(MCMT("x", 4, 4), [*N[2], *N[4]])      
    qc.append(MCMT("x", 4, 4), [*N[3], *N[5]])      
    qc.append(MCMT("x", 4, 4), [*N[4], *N[6]])      
    qc.append(MCMT("x", 4, 4), [*N[5], *N[7]])      
    qc.append(MCMT("x", 4, 4), [*N[6], *N[14]])      
    qc.append(MCMT("x", 4, 4), [*N[7], *N[15]])

    qc.barrier()

    return qc
    
# SAES().decompose().draw('mpl', reverse_bits=True)
SAES().draw('mpl', reverse_bits=True)
```

```python
# Define Hadamard Transform gate
def Htransform(qc, qubits):
    for q in qubits:
        qc.h(q)
    return qc
```

## Quantum Circuit of Grover attack on S-AES



```python
# Applying Grover attack on S-AES
k = QuantumRegister(16, "K")
w = QuantumRegister(56, "W")
o = QuantumRegister(1, "O")
c = ClassicalRegister(16, "C")
qc = QuantumCircuit(o, w, k, c)

qc.barrier()
qc.x(o)
Htransform(qc, [*k, o]) 

qc.barrier()
qc.append(SAES(), [ *w, *k])
qc.append(cipher(ciphertext), [*w[0:16]])

qc.barrier()
qc.mcx([*w[0:16]], o)
qc.append(cipher(ciphertext), [*w[0:16]])
qc.append(SAES().inverse(), [ *w, *k])

qc.barrier()
qc.append(GroverOperator(QuantumCircuit(k)),[*k])

qc.barrier()
qc.measure(k, c)

qc.draw('mpl', reverse_bits=True)
```

```python
# Initialize a Sampler backed by the stabilizer circuit simulator
exact_sampler = Sampler(backend_options=dict(method="stabilizer"))
job = exact_sampler.run(qc)
exact_quasis = job.result().quasi_dists[0]
```
